```python
# Organize data for phenome force

import pandas as pd 
import os
import glob
import random
import shutil
```

```python
data_dir = '/mnt/c/Users/Travis/Downloads/flirS11_processed_RE'
output_dir = '/mnt/c/Users/Travis/Downloads/meeting_data'
```

```python
lv1 = glob.glob(os.path.join(data_dir , '*')); lv1
plots = [os.path.basename(i) for i in glob.glob(os.path.join(data_dir, 'flirIrCamera-2020-08-04__22-57-44-695_out','plotclip_out','*'))];plots
samples = random.sample(plots, 20);samples
```

```python


for sample in samples:
    os.mkdir(os.path.join(output_dir,sample))
    for main_folder in lv1:

        try:
            date = os.path.basename(main_folder)[13:23]
            print(date)

            dest_dir = os.path.join(output_dir, sample)
            src_file = os.path.join(data_dir, main_folder, 'plotclip_out', sample, sample+'_ortho.tif')

            
            shutil.copy(src_file, dest_dir)

            dst_file = os.path.join(output_dir, sample, sample+'_ortho.tif')
            new_dst_file_name = os.path.join(output_dir, sample, f'{date}.tif')

            if os.path.exists(new_dst_file_name):
                new_dst_file_name = new_dst_file_name.replace('.tif', '_night.tif')
            os.rename(dst_file, new_dst_file_name)
        
        except:
            print(date, sample, 'was not available')


        

```
